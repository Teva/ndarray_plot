## Changing the "camera"

```python
subject = np.ones((3,4,3))
plotter = ndap.NDArrayPlotter(subject.shape)

(fig, ax) = plotter.render(subject,azim=-45,elev=25)
```

## Change element size, color, and alpha

```python
subject = np.ones((5,5,5))
plotter = ndap.NDArrayPlotter(subject.shape, scale=(0.7,0.7,0.7),color="#FF0000",alpha=1)
(fig, ax) = plotter.render(subject,azim=-45,elev=25)
```

## Selectively change colors

```python
subject = np.ones((1,10,10))
plotter = ndap.NDArrayPlotter(subject.shape, alpha=0.5)

colors = plotter.set_color("#00FF00")

i = 1
colors[0,i,i] = "#0000FF"
colors[0,i,i+1] = "#FF0000"
colors[0,i,i-1] = "#FF0000"
colors[0,i+1,i] = "#FF0000"
colors[0,i-1,i] = "#FF0000"

(fig, ax) = plotter.render(subject)
```

## Change skew, selectively change colors - Border / inner / outer

```python
reload(ndap)
subject = np.ones((5,5,5))
plotter = ndap.NDArrayPlotter(subject.shape, scale=(0.05,0.7,0.7), skewer="layered_loose")

colors = plotter.set_color("#00FF00")
alphas = plotter.set_alpha(0.6)

colors[:,:,0] = "#FF0000"
colors[:,:,-1] = "#FF0000"
colors[0,:,:] = "#FF0000"
colors[-1,:,:] = "#FF0000"
colors[:,0,:] = "#FF0000"
colors[:,-1,:] = "#FF0000"

(fig_coord, ax_coors) = plotter.render(subject)
```

## Show element coordinates and values

```python
subject = np.arange(0,9).reshape((1,3,3))
plotter = ndap.NDArrayPlotter(subject.shape, skewer="even_wide")

colors = plotter.set_color("#FFFF00")
alphas = plotter.set_alpha(0.2)

colors[:,1,:] = "#FF0000"
colors[:,0,:] = "#00FF00"
colors[:,2,:] = "#0000FF"

(fig_coord, ax_coors) = plotter.render(subject, text='coords')
(fig_values, ax_value) = plotter.render(subject, text='values')
```
